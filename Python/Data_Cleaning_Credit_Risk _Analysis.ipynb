# Fix Gender Column — Inconsistent Formats

```python
print("BEFORE FIXING GENDER")
print(df_applicants['Gender'].value_counts())
```

```python
gender_maping = {
    'Male':'Male', 
    'Female':'Female', 
    'M': 'Male', 
    'F': 'Female', 
    'MALE': 'Male', 
    'FEMALE':'Female'
}
```

```python
df_applicants['Gender'] = df_applicants['Gender'].map(gender_maping)
```

```python
print("\nAFTER FIXING GENDER")
print(df_applicants['Gender'].value_counts())
print(f"\nUnique values now : {df_applicants['Gender'].nunique()}")
```

# Fix Missing Values 

```python
print("BEFORE FIXING MISSING VALUES") 

print(f'Applicants : {df_applicants.isnull().sum().sum()} missing')
print(f'Loans : {df_loans.isnull().sum().sum()} missing')
print(f'Repayments : {df_repayments.isnull().sum().sum()} missing')
```

```python
# Annual_Income :  fill with median (income is skewed median better than mean) 


df_applicants['Annual_Income'] = df_applicants['Annual_Income'].fillna(
    df_applicants['Annual_Income'].median()
)
```

```python
# Credit_Score — fill with median


df_applicants['Credit_Score'] = df_applicants['Credit_Score'].fillna(
    df_applicants['Credit_Score'].median()
)
```

```python
# Education — fill with mode (most common value)

df_applicants['Education'] = df_applicants['Education'].fillna(
    df_applicants['Education'].mode()[0]
)
```

```python
df_applicants.isnull().sum()
```

# ── Table 2 — Loan Details ────────────────────────────────────

```python
df_loans.isnull().sum()
```

```python
df_loans['Loan_Amount'] = df_loans['Loan_Amount'].fillna(
    df_loans['Loan_Amount'].median())
```

```python
df_loans['Interest_Rate'] = df_loans['Interest_Rate'].fillna(
    df_loans['Interest_Rate'].median())
```

```python
df_loans.isnull().sum()
```

# ── Table 3 — Repayment History ───────────────────────────────

```python
df_repayments['Amount_Paid'] = df_repayments['Amount_Paid'].fillna(0)
```

```python
df_repayments['Payment_Status'] = df_repayments['Payment_Status'].fillna('Unknown')
```

```python
print("\nAFTER FIXING MISSING VALUES")
print(f"Applicants  : {df_applicants.isnull().sum().sum()} missing")
print(f"Loans       : {df_loans.isnull().sum().sum()} missing")
print(f"Repayments  : {df_repayments.isnull().sum().sum()} missing")
```

# Feature Engineering — Creating New Useful Columns

## Applicants Table

### 1. Credit Score Category

```python
df_applicants['Credit_Score_Category'] = pd.cut(
    df_applicants['Credit_Score'], 
    bins=[0,580,670,740,800,900], 
    labels= ['Poor','Fair','Good','Very Good','Excellent']
)
```

### 2. Age Group

```python
df_applicants['Age_Group'] = pd.cut(
    df_applicants['Age'], 
    bins=[20,30,40,50,65], 
    labels=['21-30','31-40','41-50','51-65']
)
```

### 3. Income Category

```python
df_applicants['Income_Category'] = pd.cut(
    df_applicants['Annual_Income'],
    bins=[0, 300000, 600000, 1000000, 2000000, 99999999],
    labels=['Low', 'Lower Middle', 'Middle', 'Upper Middle', 'High']
)
```

## Loans Table

### 1. DTI Risk Category

```python
df_loans['DTI_Risk'] = pd.cut(
    df_loans['Debt_To_Income_%'],
    bins=[0, 20, 35, 50, 999],
    labels=['Low Risk', 'Moderate Risk', 'High Risk', 'Very High Risk']
)
```

### 2. Loan Size Category

```python
df_loans['Loan_Size'] = pd.cut(
    df_loans['Loan_Amount'],
    bins=[0, 100000, 500000, 1000000, 99999999],
    labels=['Small', 'Medium', 'Large', 'Very Large']
)
```

```python
print("New columns added to Applicants:")
print(df_applicants[['Credit_Score', 'Credit_Score_Category',
                      'Age', 'Age_Group',
                      'Annual_Income', 'Income_Category']].head(5))
```

```python
print("New columns added to Loans:")
print(df_loans[['Debt_To_Income_%', 'DTI_Risk',
                'Loan_Amount', 'Loan_Size']].head(5))
```

## Fixed Duplicate Columns After Merge

When merging tables that have columns with same names
pandas automatically adds _x and _y suffixes to
distinguish them.

I found 4 such columns:
- Applicant_ID_x and Applicant_ID_y — same column

- EMI_Amount_x (from loans) and EMI_Amount_y (from repayments)
  kept repayments version and renamed it to EMI_Amount

Now all column names are clean without any suffixes.
